```python
new_df = new_df.reset_index().rename(columns={'index': 'period'})
```

```python
clean_df = new_df.iloc[2:].copy()
clean_df.head()
```

```python
clean_df['ds'] = (clean_df['period'].str.replace('INDX','').apply(lambda x: pd.to_datetime(x, format='%m%Y')))
```

# Bajra Crop

```python
prophet_df = clean_df[['ds','Bajra']].rename(columns={'Bajra': 'y'})
prophet_df.head()
```

```python
train = prophet_df[prophet_df['ds'] < '2022-01-01']
test  = prophet_df[(prophet_df['ds'] >= '2022-01-01') & (prophet_df['ds'] < '2023-01-01')]
```

```python
from prophet import Prophet

model = Prophet(
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False
)

model.fit(train)
```

```python
future = model.make_future_dataframe(periods=24, freq='MS')
forecast= model.predict(future)
```

```python
model.plot(forecast)
model.plot_components(forecast)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, root_mean_squared_error

y_true = test['y'].values
y_pred = forecast.set_index('ds').loc[test['ds'], 'yhat'].values

mape = mean_absolute_percentage_error(y_true, y_pred)
print("MAPE:", mape)
rmse = root_mean_squared_error(y_true, y_pred)
print("RMSE:", rmse)
```
